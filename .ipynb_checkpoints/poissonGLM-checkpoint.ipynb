```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import data frame
fullData = pd.read_csv("31DayData.csv")
myData = fullData[['Quarter', 'HB', 'CancerType','NumberOfEligibleReferrals31DayStandard','NumberOfEligibleReferralsTreatedWithin31Days']]
```

```python
df=myData
df.shape
# remove all cancer types from data
df = df[df['CancerType'] != 'All Cancer Types']
df
```

```python
# take a look at what we have

print(df.corr())
df.dtypes
```

```python
import seaborn as sns

sns.pairplot(df,hue='CancerType')
```

```python
#sns.boxplot(y='Count',x='Sex',data=df)
#sns.boxplot(y='Count',x='CancerType',data=df)
#sns.boxplot(y='Count',x='HB',data=df)
sns.boxplot(y='NumberOfEligibleReferrals31DayStandard',x='CancerType',data=df)

#sns.boxplot(y='Count',x='Age Group',data=df)
```

```python
# choose explanatory variables
X = df[['Quarter','HB','CancerType']]#, 'CancerType', 'Sex']]
# turn categorical into dummies
X = pd.get_dummies(data=X, drop_first=True)
X.head()
# choose output data
Y = df['NumberOfEligibleReferrals31DayStandard']
Y.head()
```

```python
# build model
from sklearn.linear_model import PoissonRegressor
pr = PoissonRegressor(alpha=0, fit_intercept=True)


# fit model
pr.fit(X, Y)
pr.get_params()
print(pr.intercept_)
coeff_parameter = pd.DataFrame(pr.coef_,X.columns,columns=['Coefficient'])
coeff_parameter
```

```python
#trying to use it to predict (from https://medium.com/analytics-vidhya/implementing-linear-regression-using-sklearn-76264a3c073c)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=101)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
pr.fit(X_train,y_train)
predictions = pr.predict(X_test)
predictions
```

```python
sns.regplot(y_test,predictions)
```

```python
import statsmodels.api as sm
X_train_Sm= sm.add_constant(X_train)
X_train_Sm= sm.add_constant(X_train)
ls=sm.OLS(y_train,X_train_Sm).fit()
print(ls.summary())
```
